```python
from Bio import AlignIO
from collections import Counter

alignment = AlignIO.read("../results/alignments/aligned.fasta", "fasta")
print(alignment)

# Consensus sequence using modern Biopython API
alignment_obj = alignment.alignment
consensus_list = []
for pos in range(alignment_obj.shape[1]):
    col = alignment_obj[:, pos]
    counts = Counter(col)
    if counts:
        res = counts.most_common(1)[0][0]
        if res != '-':
            consensus_list.append(res)
consensus = ''.join(consensus_list)
print("Consensus:", consensus)
```
