```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
from Bio.Align import substitution_matrices

# Load substitution matrix
matrix = substitution_matrices.load("BLOSUM62")

# Example protein sequences
seq1 = "MKTAYIAKQRQISFVKSHFSRQLEERLGLIEVQ"
seq2 = "MKALYIAKQRQISFVKSHFSRQLDEELGLIEVQ"

print("=== GLOBAL ALIGNMENT (Needleman-Wunsch) ===")
global_alignments = pairwise2.align.globalds(seq1, seq2, matrix, -10, -0.5)

for alignment in global_alignments[:1]:
    print(format_alignment(*alignment))

print("\n=== LOCAL ALIGNMENT (Smith-Waterman) ===")
local_alignments = pairwise2.align.localds(seq1, seq2, matrix, -10, -0.5)

for alignment in local_alignments[:1]:
    print(format_alignment(*alignment))
```
